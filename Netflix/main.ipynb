```python
df.fillna({"director":df["director"].mode()[0], "cast":df["cast"].mode()[0], "country":df["country"].mode()[0], "date_added":df["date_added"].mode()[0], "rating":df["rating"].mode()[0], "duration":df["duration"].mode()[0]}, inplace=True)
```

```python
df["country"]=df["country"].str.split(",")
filter_country = df["country"].apply(lambda x: x[0]).value_counts().head(10)
```

```python
plt.barh(filter_country.index, filter_country.values)
plt.title("Top 10 Countries with Most Content on Netflix")
plt.xlabel("Number of movies")
plt.ylabel("Country")
plt.show()
```

```python
genres=df["genres"].str.split(",").explode("genres")
count= genres.value_counts().head(10)
sns.barplot(x=count.values, y=count.index)
plt.xticks(rotation=90)
plt.title("Top 10 Genres on Netflix")
plt.xlabel("Number of movies")
plt.ylabel("Genre")
plt.show()
```

```python
get_minutes = df[df["type"] == "Movie"]["duration"]
plt.hist(get_minutes, bins=20)
plt.title("Distribution of Movie Durations on Netflix")
plt.xlabel("Duration (minutes)")
plt.ylabel("count")
plt.show()
```

```python
get_minutes = df[df["type"] == "TV Show"]["duration"]
plt.hist(get_minutes, bins=20)
plt.title("Distribution of TV Show Durations on Netflix")
plt.xlabel("Duration (minutes)")
plt.ylabel("count")
plt.show()
```

```python
years= df["release_year"].value_counts().head(10)
sns.barplot(x=years.index, y=years.values)
plt.title("Top 10 Release Years on Netflix")
plt.xlabel("Number of movies")
plt.ylabel("Release Year")
plt.show()
```

```python
# Convert duration column to string first
df["duration"] = df["duration"].astype(str)
df["duration_num"] = df["duration"].str.extract('(\d+)').astype(float)
# Encode rating categories into numbers
df["rating_encoded"] = pd.factorize(df["rating"])[0]

# Top 10 genres
top_genres = df["genres"].value_counts().head(10).index

# Filter dataset
top_df = df[df["genres"].isin(top_genres)]

# Group data
heatmap_data = top_df.groupby("genres").agg({
    "rating_encoded": "mean",
    "duration_num": "mean"
})

# Heatmap
plt.figure(figsize=(10,6))

sns.heatmap(
    heatmap_data,
    annot=True,
    cmap="coolwarm",
    linewidths=1
)

plt.title("Heatmap of Genres, Ratings and Duration")

plt.show()
```
